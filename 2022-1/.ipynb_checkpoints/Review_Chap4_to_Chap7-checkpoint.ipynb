> 데이터프레임 - DataFrame

```python
# 관련 라이브러리 불러오기
import pandas as pd           #판다스
import matplotlib as mpl       #맷플롭립
import matplotlib.pylab as plt   #맷플롭립
import seaborn as sns          #시본 
import numpy as np           #넘파이

#데이터 준비 시본에 내장된 팁 데이터 
tips = sns.load_dataset("tips")    # 팁 데이터
tips

# 기본정보확인 - tips.info()
tips.info()

# head, tail 확인
tips.head()
tips.tail()
```

```python
# index 접근
# iloc / loc
# tips.drop([2]) tips 데이터의 2번 index 삭제
# 마지막행 읽어오기 tips.iloc[-1]
print("iloc", tips.iloc[2]) # 행(rows) 번호
print("loc", tips.loc[2]) # 인덱스(index) 번호
print("multi-loc", tips.loc[[1,2,3]])
```

```python
# 행읽어오기, 열읽어오기
tips[0:3] # 0~2행까지 가져오기 [총 3개의 행을 가져옴]
tips.iloc[0:3] # 0~2 행가져오기
tips.loc[0:3] # 0~3 index에 해당하는 행 가져오기
tips.loc[0:3,['sex','day']] #0~3 index에서 열이 sex, day만 가져오기
# slicing
tips.iloc[1:3,:] # 1부터 2행까지 모든 열을 가져옴
tips.iloc[:,1:3] # 1부터 2열까지 모든 행을 가져옴
```

```python
# 단일 조건
tips[tips.tip > 5] # tip이 5이상
tips[tips["tip"] > 5] # 위와 동일
# 다중 조건 : and 조건 – 남자 손님이면서 비흡연자
tips[(tips['sex']=='Male') & (tips['smoker']=='No')]
# 특정조건 isin
tips['day'].isin(['Sun']) # True
tips[tips['tip'].isin([1])] # tips이 1인 행을 가져옴
```

```python
# 열 변수 추출 - 열추출
tips.tip # tips['tip'] 와 동일, tip 열변수 추출
tips [['total_bill','tip','day']] # total_bill, tip, day 열 추출 : 여러개의 열을 추출하고자 할 경우 [[]] 사용
tips['total'] = tips['total_bill'] + tips['tip'] # 파생변수 만들기
```

```python
# 데이터 확인
tips.count( ) # 갯수
len(tips) # 크기
tips.index # index 보기
tips.columns # 열 보기
tips.values # 열구조 보기
# 데이터 정렬
tips.sort_values(by=['total_bill'], axis=0) # 열기준으로 axis = 0 오름차순 정렬, ascending=False 추가하면 내림차순으로
tips.sort_index(axis=1) # 알파벳순 정렬 axis = 1
# 열합계, 행합계
tips.sum() # 열합계
tips.sum(axis=1) # 행합계
```

```python
import pandas as pd

# 병합용 데이터프레임 생성
df1 = pd.DataFrame({ '고객번호': [1001, 1002, 1003, 1004, 1005, 1006, 1007],
    '이름': ['강감찬', '홍길동', '이순신', '장보고', '유관순', '신사임당', '세종대왕']
}, columns=['고객번호', '이름'])
df2 = pd.DataFrame({'고객번호': [1001, 1001, 1005, 1006, 1008, 1001],
    '금액': [10000, 20000, 15000, 5000, 100000, 30000]
}, columns=['고객번호', '금액'])

# join (병합)
pd.merge(df1, df2) # inner join
pd.merge(df1, df2, how='outer') # outer join
pd.merge(df1, df2, how='left') # left join
pd.merge(df1, df2, how='right') # right join
```

> 데이터 재구조화

```python
# cut, qcut()
df1=pd.cut(df.math, 3) #math 변수에 대해서 3개 동일한 길이로 범주형 변수 : 범주 구간은 Categories (3, interval[float64]): [(19.93, 43.333] < (43.333, 66.667] < (66.667, 90.0]] 로 3개 구간 길이가 동일함
print(df1) 
df2= pd.qcut(df.math, 3, labels=False) # math에 대해서 각 범주 구간 별로 동일하게 3개의 개수 : labels=False 로 설정하여 label이 0, 1, 2 구간을 0부터 순차적으로 1씩 증가
print(df2)
df3=df2.agg(['count', 'mean', 'std', 'min', 'max']) # 그룹별 통계량
print(df3)
```

```python
# 원핫코딩 : 범주형일 경우만 해당 
df = pd.get_dummies(df) # 범주형 원핫코딩
print(df)
# 데이터 전치 : 위치 바꿈 : .T
print(df.T)
```

```python
# pivot
print(pd.pivot_table(df,index='class', columns='sex', values='science')) # 반(class) 변수를 행 데이터로 피봇
print(pd.pivot_table(df,index='sex', columns='class', values='science')) # 성별(sex) 변수를 행 데이터로 피봇
pd.melt(df,id_vars=['student_no'], value_vars=['class']) # melting
print(df)
```

```python
#판다스 라이브러리 불러오기
import pandas as pd 
#파일경로를 찾고 변수 file_path에 저장
file_path= 'C:/Rtest/timeseries.csv'
#read csv()함수로 데이터프레임 변환 
df= pd.read_csv(file_path)

# 현재 날짜를 나타내는 Date컬럼은 문자형(object)이므로 to_datetime()함수를 이용해서 Date컬럼을 시계열 객체(Timestamp)로 변환
df.info()
df['new_Date'] = pd.to_datetime(df['Date'])
print(df.head())
print('\n')
```

```python
# Timestamp와 Period의 차이 확인 , Timestamp를 Period로 변환
# Period객체는 to_period(freq='기간인수')를 통해 datetime변수에 대해 어떤 기간에 따른 자료형을 생성하고자 할 때 주로 활용된다. 아무 자료형이나 사용할 수 없고 datetime유형에 대해서만 적용 가능
dates = ['2020-01-01','2020-03-01','2021-09-01']
ts_dates = pd.to_datetime(dates)
print(ts_dates)
print('\n')
# Timestamp를 Period변환
pr_day = ts_dates.to_period(freq='D')   #1일 기간
print(pr_day)
pr_month = ts_dates.to_period(freq='M') #1개월 기간
print(pr_month)
pr_year = ts_dates.to_period(freq='A')  #1년 기간
print(pr_year)
```

> 시계열활용.시각화

```python
# 시계열 활용
# 날짜 데이터 분리 날짜 데이터 분리 : dt.year, dt.month, dt.day
#판다스 라이브러리 불러오기

import pandas as pd 
#파일경로를 찾고 변수 file_path에 저장
file_path= 'C:/Rtest/timeseries.csv'
#read csv()함수로 데이터프레임 변환 
df= pd.read_csv(file_path)
df.info()
df['new_Date'] = pd.to_datetime(df['Date'])
print(df.head())
print('\n')
 
df['Year'] = df['new_Date'].dt.year
df['Month'] = df['new_Date'].dt.month
df['Day'] = df['new_Date'].dt.day
df.head()
```

```python
# to_period()
df['Date_yr'] = df['new_Date'].dt.to_period(freq = 'A')  # 연도까지
df['Date_m'] = df['new_Date'].dt.to_period(freq = 'M')   # 연월까지
print(df.head())
# 날짜 인덱스 지정
df.set_index('new_Date',inplace=True)
df.head()
```

```python
# 날짜 인덱싱
print(df.loc['2015-07']) # 7월에 해당하는 row 인덱싱
print('\n')
print(df['2019-06']) # 6월에 해당하는 row 인덱싱
print('\n')
print(df['2018-06-25' : '2018-06-20'])  # 해당기간의 인덱싱
```
