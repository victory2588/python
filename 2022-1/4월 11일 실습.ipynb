```python
import pandas as pd
import matplotlib as mpl #맷플롭립
import matplotlib.pylab as plt #맷플롭립
import seaborn as sns #시본
import numpy as np #넘파이

#시리즈를 데이터프레임의 새로운 행으로 연결
s1=pd.Series([0,1],index=["A","B"])
s2=pd.Series([2,3,4],index=["c","d","e"])
print(s1)
print(s2)
print(pd.concat([s1,s2]))

#열 기준 데이터 연결 -> axis=1 사용
df1 = pd.DataFrame([['Dog','3'], ['Bird','10'], ['Tiger','6'], ['Moose','3']],index = ['0','1','2','3'],columns = ['동물','나이'])
df1
df2 = pd.DataFrame([['집','0'], ['초원','0'], ['수풀','0'], ['초원','1']],index = ['0','1','2','3'],columns = ['사는곳','뿔의 개수'])
df2
pd.concat([df1, df2], axis=1) 

#데이터 병합
df1 = pd.DataFrame({ '고객번호': [1001, 1002, 1003, 1004, 1005, 1006, 1007],
                    '이름': ['강감찬', '홍길동', '이순신', '장보고', '유관순', '신사임당', '세종대왕']}
                   , columns=['고객번호', '이름'])
print(df1)
df2 = pd.DataFrame({'고객번호': [1001, 1001, 1005, 1006, 1008, 1001],
                    '금액': [10000, 20000, 15000, 5000, 100000, 30000]}
                   ,columns=['고객번호', '금액'])
print(df2)

pd.merge(df1,df2) #inner join
pd.merge(df1,df2,how="outer") #full join
pd.merge(df1, df2, how='left')
pd.merge(df1, df2, how='right') #한 쪽을 기준으로 join

#범주형 데이터
tips=sns.load_dataset('tips')
tips
tips['sex'].value_counts()
tips['day'].value_counts()
pd.crosstab(tips['sex'],tips['day'])
pd.crosstab(tips['sex'],tips['day'], margins=True)
#pd.crosstab(tips['sex'],tips['day'].apply(lambda r: r/len(tips),axis=1) 오류난다

#결측값 처리
file_path='C:/Users/victo/anaconda3/python/실습데이터/bicycle.csv'

df=pd.read_csv(file_path,engine='python')
df

df.isnull()

df.isnull().sum(1) #행단위 결측값 개수
df.notnull().sum(1) #행단위 실제값 개수

df_drop_allrow=df.dropna(axis=0) #결측치가 있는 전체 행 제거
df_drop_allrow
df_drop_allcolumns=df.dropna(axis=1) #결측치가 있는 전체 열 제거
df_drop_allcolumns

#특정 행 또는 열 결측치 제거, 대여소번호 컬럼 제거(비교확인)
df['대여소번호'].dropna()

# 하단 df.dropna(axis=0)와 동일
df[['대여소번호','대여거치대','이용시간']].dropna()
df[['대여소번호','대여거치대','이용시간']].dropna(axis=0)
df[['대여소번호','대여거치대','이용시간']].dropna(axis=1)

df_1 = df.fillna(0) #결측값을 특정 값(0)으로 대체
df_1

df_2 = df.대여소번호.fillna(0) #특정 컬럼 결측값을 특정 값(0)으로 대체
df_2

df_3 = df.fillna('missing') #결측값을 문자열('missing')로 대체
df_3

df_4=df.fillna(df.mean()) # 각 칼럼의 평균을 구해서 대응하는 칼럼의 결측값을 대체하는 방법
df_4

df_5= df.mean()['이용거리'] # 특정항목 평균 구하기(이용거리평균)
df_5

df_6 = df.fillna(df.mean()['이용거리']) #특정항목 평균으로 대체
df_6

df_7 = df.이용거리.fillna(df.mean()['이용거리'])
df_7
```
