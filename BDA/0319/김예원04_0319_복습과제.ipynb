### cross_val_score
- kfold, skfold 등을 쉽게 사용할 수 있다.
- 분류 문제인지 회귀 문제인지에 따라서 저절로 더 좋은 교차검증 방법 사용
- cross_val_score(사용할 모델, X(데이터), Y(데이터), scoring(평가할 스코어링 방법), cv 몇 번 교차검증 할 것인가?, n_jobs(cpu사용대한 지정 -1이면 모든 cpu), verbose(진행상황에 대한 유무 보기), fit_params 등등 )
- 대부분이 이진 분류 기준!

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score, cross_validate
from sklearn.datasets import load_iris
import numpy as np
```

```python
iris=load_iris()
dt_clf=DecisionTreeClassifier(random_state=111)

data=iris.data
target=iris.target
```

```python
# 정확도
scores=cross_val_score(dt_clf,data,target,scoring='accuracy',cv=10)
print('정확도: ',np.round(scores,4))
print('정확도 평균: ',np.round(np.mean(scores),4))

# 정밀도
scores=cross_val_score(dt_clf,data,target,scoring='precision_macro',cv=10)
print('정밀도: ',np.round(scores,4))
print('정밀도 평균: ',np.round(np.mean(scores),4))

# 재현율
scores=cross_val_score(dt_clf,data,target,scoring='recall_macro',cv=10)
print('재현율: ',np.round(scores,4))
print('재현율 평균: ',np.round(np.mean(scores),4))

# f1 스코어
scores=cross_val_score(dt_clf,data,target,scoring='f1_macro',cv=10)
print('f1: ',np.round(scores,4))
print('f1 평균: ',np.round(np.mean(scores),4))
```

```python
import seaborn as sns
titanic=sns.load_dataset('titanic')

titanic
```

```python
titanic_data=titanic[['survived','pclass','age','sibsp','parch','fare']]    # 분석에 필요한 컬럼만 뽑은 후

tt_x=titanic_data[['pclass','sibsp','parch','fare']]
tt_y=titanic_data['survived']

tt_scores=cross_val_score(dt_clf,tt_x,tt_y,scoring='accuracy',cv=10)
print('정확도',np.round(tt_scores,4))
print(np.round(np.mean(scores),4))

tt_scores=cross_val_score(dt_clf,tt_x,tt_y,scoring='precision',cv=10)
print('정밀도',np.round(tt_scores,4))
print(np.round(np.mean(scores),4))

tt_scores=cross_val_score(dt_clf,tt_x,tt_y,scoring='f1',cv=10)
print('f1',np.round(tt_scores,4))
print(np.round(np.mean(scores),4))

tt_scores=cross_val_score(dt_clf,tt_x,tt_y,scoring='recall',cv=10)
print('재현율',np.round(tt_scores,4))
print(np.round(np.mean(scores),4))
```

### cross_validate: 소요시간, train과 test에 대한 모든 정확도 점수

```python
cv_scores=cross_validate(dt_clf,tt_x,tt_y,scoring='accuracy',cv=10,return_train_score=True)
print(cv_scores)

cv_scores=cross_validate(dt_clf,tt_x,tt_y,scoring='precision',cv=10,return_train_score=True)
print(cv_scores)
```

### 3가지 모델을 이용한 평가
- DecisionTreeClassifier(),RandomForestClassifier(),LogisticRegression()

```python
model=[DecisionTreeClassifier(),RandomForestClassifier(),LogisticRegression()]
name=['DT','RF','LR']

for model, name in zip(model,name):
    print('###############사용한 알고리즘',name,'#####################')
    for score in ['accuracy','precision','recall','f1']:
        print(score)
        print('-')
        print(cross_val_score(model,tt_x,tt_y,scoring=score,cv=10))
```
