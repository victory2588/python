### 한글 외 문자 제거

```python
import re

movie['document'] = movie['document'].apply(lambda x : re.sub(r'[^ ㄱ-ㅣ가-힣]+', " ", x))
movie.head(10)
```

```python
text = movie['document']
label = movie['label']
```

## 데이터 셋 분리

```python
from sklearn.model_selection import train_test_split

train_x, test_x, train_y, test_y = train_test_split(text, label , test_size=0.2, random_state=0)
print(len(train_x), len(train_y), len(test_x), len(test_y))
```

## 토큰화 및 TF-IDF 벡터화
- 토큰화: 주어진 코퍼스(corpus)에서 토큰(token)이라 불리는 단위로 나누는 것
- 토큰화 시 주의사항
1. 구두점이나 특수문자를 단순 제외해도 무방한지 확인 (ex, Ph.D, AT&T, $2,000, 11/22/1990, 45.34)
2. 줄임말과 단어 내 띄어쓰기 (we’re, thx, 고대, 영끌)
3. 한국어는 어절 독립단어가 아니므로 형태소(자립, 의존) 분해가 필요
4. 한국어는 띄어쓰기가 지켜지지 않고 있음

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from konlpy.tag import Okt

okt = Okt()

tfv = TfidfVectorizer(tokenizer=okt.morphs, ngram_range=(1,2), min_df=3, max_df=0.9)
tfv.fit(train_x)
tfv_train_x = tfv.transform(train_x)
print(tfv_train_x)
```

## 모델 구축
- penalty, max_iter, C, solver 를 넣어 최적의 파라미터 찾기

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV    #하이퍼 파라미터 최적화

clf = LogisticRegression(random_state=0,)
params = {'penalty':['l2','l1'],'max_iter':[100,500],'C': [1,3,5,10,15, 18, 19, 20],'solver':['lbfgs','liblinear','newton-cg']}
grid_cv = GridSearchCV(clf, param_grid=params, cv=3, scoring='accuracy', verbose=1)
grid_cv.fit(tfv_train_x, train_y)

# 최적의 평가 파라미터는 grid_cv.best_estimator_에 저장됨
print(grid_cv.best_params_, grid_cv.best_score_)# 가장 적합한 파라메터, 최고 정확도 확인
```

## 분석 모델 평가
- 평가용 데이터의 특징 벡터화

```python
from sklearn.metrics import accuracy_score

tfv_test_x = tfv.transform(test_x)
# test_predict = grid_cv.best_estimator_.score(tfv_test_x,test_y)
test_predict = grid_cv.best_estimator_.predict(tfv_test_x)

print('감성 분류 모델의 정확도 : ',round(accuracy_score(test_y, test_predict), 3))
```

```python
input_text = '역시 레전드..'
#입력 텍스트에 대한 전처리 수행
input_text = re.compile(r'[ㄱ-ㅣ가-힣]+').findall(input_text)
input_text = [" ".join(input_text)]
# 입력 텍스트의 피처 벡터화
st_tfidf = tfv.transform(input_text)

# 최적 감성 분석 모델에 적용하여 감성 분석 평가
st_predict = grid_cv.best_estimator_.predict(st_tfidf)

#예측 결과 출력
if(st_predict == 0):
    print('예측 결과: ->> 부정 감성')
else :
    print('예측 결과: ->> 긍정 감성')
```
