## 정밀도/재현율은 trade-off 관계
- 재현율이 중요한 경우: 실제 positive 양성 데이터를 Negative 잘못 판단하게 되면 문제가 되는 경우 -> 질병 진단 시, 사기 판단 여부
- 정밀도가 중요한 경우: 재현율의 상황과 반대, 음성 데이터를 양성이라고 잘못 판단하는 경우 -> 스팸 메일 분류 시

>문제 상황에 대한 특성을 파악하는 것이 중요

## F1 스코어

- 정밀도와 재현율을 결합한 지표
- 어느 한 쪽으로 치우치지 않게 하기 위해 사용
- 0.0 ~ 1.0 사이의 값을 가지며, 높을 수록 좋음
- Thresholds 을 이용해서 어느 순간에 P 또는 N 으로 구별할 지 정할 수 있다.

```python
# 직접 계산해보기
def f_1(pr,re):
    print(2*((pr*re)/(pr+re)))

# pr 0.9 re 0.1 / pr 0.5 re 0.5 중 어떤 것이 더 성능이 좋은가?

f_1(0.5,0.5)
f_1(0.9,0.1)
```

>정밀도, 재현율 둘 다가 가장 높을 수 있는 상황이 best

## ROC 곡선과 AUC

- 이진 분류/예측에서 사용
- FPR x축 , TPR Y 축 FPR변화에 따른 TRP의 변화 곡선을 나타낸 것
- 현 위의 점의 의미: threshold가 변함에 따라서 False Positive Rate(FPR)과 True Positive Rate(TPR)의 값이 바뀐다는 것
- 현의 휨 정도의 의미: 두 클래스를 더 잘 구분할 수 있게 된다면, 커브는 좌상단에 붙게 된다.

- ROC곡선 자체는 FPR, TPR 변화의 값을 보는 데 이용하고, 분류의 성능 지표로 사용되는 건 ROC 곡선 면적에 기반한 AUC 값
- AUC값(면적 값)이 1에 가까울 수록 좋은 수치
