### K-means 
- k개를 지정하여 평균을 통해 군집화한다는 뜻으로 생각해 보자!
- 비지도학습이라는 게 정답을 준 것이 아니라 정답이 없는 데이터에서 머신에게 패턴을 찾아달라고 하는 것
- y 라벨이 없는 상태에서 데이터를 집어 넣고 특성을 군집화하는 것
- 군집화는 같은 패턴끼리 묶는다.
- 작동원리가 어떤식으로 진행하는지는 '평균' 
- 거리의 기반으로 작동을한다. (유클리디안 거리)
- k -means의 k는 몇 개의 클러스터(군집화)를 할 것인가?
- k는 분석가가 초기값을 입력해야 한다. 예를 들어 3개라고 하면 3개의 군집 포인트가 데이터의 패턴을 거리기반(평균) 군집을 만든다.
- 초기값을 지정해 줘야함
- 단점은 둘 중 거리기반으로만 하기 때문에 거리가 55,45 비슷한 거리 차이어도 55가깝게 하드클러스터링을 하기 때문에
- 이 부분이 문제가 된다.
- 거리기반이라 이상치에 민감해지고
- 단순히 거리기반으로 평균으로 진행해서 그룹자체의 분산 구조는 반영하지 못한다.
- 데이터 양이 많아서 반복횟수가 많아지면 속도가 느린 부분도 있음
- 몇 개의 군집을 선택해야 하는지 가이드를 주는 것이 어렵다.

- 하지만 좋은점은
- 군집화하기에 가장 많이 활용된다.
- 이해도가 쉽다. 알고리즘 간결하다. 거리 기반으로 평균으로 이해하면 되니깐.

### 거리기반으로 이야기했으면 
- 거리기반이기 때문에 값에 대한 차이가 민감할 것
- 클러스터링할 때 스케일링을 해야 한다.
- 스케일링을 해보자!

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler
```

```python
df_sp_1
```

```python
st_scaler = StandardScaler()
```

```python
st_scaler.fit(df_sp_1)
```

```python
df_sp_1_sc = st_scaler.transform(df_sp_1)
```

```python
df_sp_1_sc
```

```python
df_tt_sc= pd.DataFrame(df_sp_1_sc, columns = df_sp_1.columns)
```

```python
df_tt_sc
```

```python
kmeans.fit(df_tt_sc)
```

```python
#실제로 군집이 어떻게 되었나?
print(kmeans.labels_) # 자동으로 군집화 된 내용을 공유해 준다.
```

```python
df_tt_sc['cluster'] = kmeans.labels_
```

```python
df_tt_sc.groupby(['cluster'])['ash'].count()
```

- 거리기반으로 민감하게 움직이기 때문에 스케일링과 변수들의 이상치 등을 고려하여 군집을 만들어야 합니다.

- 몇 개의 군집이 정말 적정한것인가? 
- 질문에 대한 답변으로 바라보는 점수들이 있다.
- inertia, 실루엣 점수, 엘보우 점수 군집의 점수를 판단하는 것인데
- 점수가 잘 나온다고해서 해당 군집의 개수를 선정하는 것이 아니라
- 정말 군집이 잘 형성이 되었는지 시각화를 통해서도 봐야 한다.


```python
df_tt_sc
```

- 4차원 이상의 차원을 시각화하게 되면 우리는 이 부분을 이해할 수 없다.
- 이런 다차원의 특성을 확인하기 위해서 차원축소를 하는 것이다. pca 
- 차원 축소를 하는 이유가 고차원에 대한 변수들을 군집화 했을 때 시각화하면 볼 수가 없다.
- 이것을 2개의 차원으로 축소하여 군집화를 시각화해보면서 정말 잘 시각화되었는지 확인하는 것
