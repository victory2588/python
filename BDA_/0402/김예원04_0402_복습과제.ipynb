### KNN 알고리즘, Kmeans

### K-means 
- k개를 지정하여 평균을 통해 군집화한다는 뜻으로 생각해 보자!
- 작동원리: 평균, 거리 기반(유클리디안) 
- 초기값을 지정해 줘야함, 몇 개의 군집을 선택해야 하는지 가이드를 주는 것이 어려움
- 단점: 거리 기반이기 때문에 나타나는 문제(이상치에 민감, 그룹 자체의 분산 반영X), 속도 느림
- 장점: 군집화에 가장 많이 활용, 이해 쉬움, 알고리즘 간결


```python
# 클래스 별 갯수 확인
wine['class'].value_counts()
```

### KMeans 사용



```python
kmeans=KMeans(n_clusters=3,init='k-means++',max_iter=300,random_state=111)
kmeans
```

```python
wine_x=wine.drop('class',axis=1)
wine_x
```

```python
kmeans.fit(wine_x)
print(kmeans.labels_)
```

```python
wine['cluster']=kmeans.labels_
wine
```

```python
wine.groupby(['class','cluster'])['alcohol'].count()
```

- 클래스 값과 비교했을 때 군집화가 잘 이루어지지 않음
- 변수가 많은 고차원 상태로 분석을 진행한 상태
- 중요한 변수만을 골라 다시 돌리는 것이 필요

### 변수 선택
- 클래스 별로 평균 차이가 큰 경우 이상치가 있을 것이라 판단 후 제외

```python
col=list(wine.columns)

for i in col:
  print('-----------------',i ,'mean 비교')
  print(wine.groupby(['class'])[i].mean())
```

- alcohol/ash/nonflavanoid_phenols/proanthocyanins 선택 

```python
wine_xx=wine[['alcohol', 'ash', 'nonflavanoid_phenols' ,'proanthocyanins']]

kmeans.fit(wine_xx)
print(kmeans.labels_)
```

```python
wine['cluster2']=kmeans.labels_
wine.groupby(['class','cluster2'])['alcohol'].count()
```

### 스케일링
- 거리 기반이기 때문에 값 차이에 민감
- 클러스터링 할 때 스케일링을 꼭 해야함!

```python
from sklearn.preprocessing import StandardScaler,MinMaxScaler,RobustScaler

wine_xx
```

```python
st_scaler=StandardScaler()
st_scaler.fit(wine_xx)
```

```python
wine_sc=st_scaler.transform(wine_xx)
wine_sc
```

```python
wine_tt_sc=pd.DataFrame(wine_sc,columns=wine_xx.columns)    # 스케일링 된 값을 데이터프레임 형식으로 만듦
wine_tt_sc
```

```python
kmeans.fit(wine_tt_sc)
wine['cluster3']=kmeans.labels_
wine.groupby(['class','cluster3'])['ash'].count()
```
