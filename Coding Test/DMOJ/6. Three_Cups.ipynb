### 내 풀이

```python
move=input()
ball=1

for i in range(len(move)):
    if move[i]=='A':
        if ball==1:
            ball=2
        elif ball==2:
            ball=1
    elif move[i]=='B':
        if ball==2:
            ball=3
        elif ball==3:
            ball=2
    else:
        if ball==3:
            ball=1
        elif ball==1:
            ball=3
print(ball)
```
