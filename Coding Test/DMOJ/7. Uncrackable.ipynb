```python
password=input()

lower=0
upper=0
digit=0

if len(password)<8 or len(password)>12:
    print('Invalid')
else:
    for i in range(len(password)):
        if 'a'<=password[i]<='z':
            lower+=1
        elif 'A'<=password[i]<='Z':
            upper+=1
        elif 0<=int(password[i])<=9:
            digit+=1
    if lower>=3 and upper>=2 and digit>=1:
        print('Valid')
    else:
        print('Invalid')
```
