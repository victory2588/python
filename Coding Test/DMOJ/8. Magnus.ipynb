```python
honi=input()

honiBlock=[]
honicnt=0

for i in range(len(honi)):
    if honi[i]=="H":
        honiBlock.clear()
        honiBlock.append(honi[i])
    elif honi[i]=="O":
        if "O" in honiBlock:
            continue
        honiBlock.append(honi[i])
    elif honi[i]=="N":
        if "N" in honiBlock:
            continue
        honiBlock.append(honi[i])
    elif honi[i]=="I":
        honiBlock.append(honi[i])
        if 'HONI' in ''.join(honiBlock):
            honicnt+=1
            honiBlock.clear()
        if "I" in honiBlock:
            honiBlock.clear()
            continue

print(honicnt)
```
