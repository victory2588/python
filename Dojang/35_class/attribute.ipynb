### 클래스 속성과 인스턴스 속성
- 클래스 속성: 클래스에서 바로 속성을 만듦. 클래스에서 만드니까 모든 인스턴스에서 공유
- 인스턴스 속성: __init__ 메소드에서 만듦. 인스턴스별로 독립되어 있음. 각 인스턴스가 값을 따로 저장해야 할 때 사용

```python
# 클래스 속성 사용하기
class Person:
    bag=[]

    def put_bag(self,stuff):
        self.bag.append(stuff)

james=Person()
james.put_bag('책')

maria=Person()
maria.put_bag('열쇠')

print(james.bag)
print(maria.bag)  # 넣은 물건들이 합쳐져서 나옴
```

```python
# 가방을 여러 사람이 공유하지 않으려면? -> bag를 인스턴스 속성으로 만들기

class Person:
    def __init__(self):
        self.bag = []
 
    def put_bag(self, stuff):
        self.bag.append(stuff)
 
james = Person()
james.put_bag('책')
 
maria = Person()
maria.put_bag('열쇠')
 
print(james.bag)
print(maria.bag)
```

```python
# 비공개 클래스 속성 생성
class Knight:
    __item_limit = 10    # 비공개 클래스 속성
 
    def print_item_limit(self):
        print(Knight.__item_limit)    # 클래스 안에서만 접근할 수 있음
 
 
x = Knight()
x.print_item_limit()    # 10
 
#print(Knight.__item_limit)    # 클래스 바깥에서는 접근할 수 없음
```

### 클래스, 메소드에서 독스트링 사용하기

```python
class Person:
    '''사람 클래스'''
    def greeting(self):
        '''인사 메서드입니다.'''
        print('Hello')
 
print(Person.__doc__)             # 사람 클래스입니다.
print(Person.greeting.__doc__)    # 인사 메서드입니다.
 
maria = Person()
print(maria.greeting.__doc__)     # 인사 메서드입니다.
```
