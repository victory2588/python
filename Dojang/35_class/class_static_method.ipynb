### 메소드의 종류
- 정적 메소드: @staticmethod, 클래스에서 바로 메소드 호출 가능, 인스턴스 속성과 메소드가 필요 없을 때 사용
- 클래스 메소드: @classmethod

- 정적 메소드


```python
class Calc:
    @staticmethod
    def add(a,b):
        print(a+b)
    
    @staticmethod
    def mul(a,b):
        print(a*b)

Calc.add(10,20)  # 인스턴스 생성해서 사용할 필요 없음
Calc.mul(10,20)
```

- 클래스 메소드


```python
class Person:
    count=0

    def __init__(self):
        Person.count+=1  # 클래스 속성에 접근한다고 정의
    
    @classmethod         # 메소드 안에서 클래스 속성, 클래스 메소드에 접근해야할 때 사용
    def print_count(cls):
        print('{0}명 생성되었습니다.'.format(cls.count))

james = Person()
maria = Person()
 
Person.print_count()
```
