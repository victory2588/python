```python
class Date:
    @staticmethod
    def is_date_valid(date_string):
        year,month,date=map(int,date_string.split('-'))  # -로 구분된 데이터니까 split('-')
        return month<=12 and date<=31                    # map(함수, 요소)

                                                            
if Date.is_date_valid('2000-10-31'):
    print('올바른 날짜 형식입니다.')
else:
    print('잘못된 날짜 형식입니다.')
```
