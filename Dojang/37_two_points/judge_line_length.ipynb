```python
import math
 
class Point2D:
    def __init__(self, x=0, y=0):
        self.x = x
        self.y = y
 
length = 0.0
p = [Point2D(), Point2D(), Point2D(), Point2D()]
p[0].x, p[0].y, p[1].x, p[1].y, p[2].x, p[2].y, p[3].x, p[3].y = map(int, input().split())
a1,a2,a3=p[1].x-p[0].x,p[2].x-p[1].x,p[3].x-p[2].x
b1,b2,b3=p[1].y-p[0].y,p[2].y-p[1].y,p[3].y-p[2].y
c1,c2,c3=math.sqrt((a1*a1)+(b1*b1)),math.sqrt((a2*a2)+(b2*b2)),math.sqrt((a3*a3)+(b3*b3))
length=c1+c2+c3

print(length)
```

```python
import math
 
class Point2D:
    def __init__(self, x=0, y=0):
        self.x = x
        self.y = y
 
length = 0.0
p = [Point2D(), Point2D(), Point2D(), Point2D()]
p[0].x, p[0].y, p[1].x, p[1].y, p[2].x, p[2].y, p[3].x, p[3].y = map(int, input().split())
for i in range(3):
    w=p[i+1].x-p[i].x
    h=p[i+1].y-p[i].y
    length+=math.sqrt((w*w)+(h*h))
print(length)
```
