```python
class NotPalindromeError(Exception):
    def __init__(self):
        super().__init__(' .')  #   
def palindrome(word):
    if word!=''.join(reversed(word)):
        # word!=word[::-1]
        raise NotPalindromeError
    else:
        print(word)
try:
    word = input()
    palindrome(word)
except NotPalindromeError as e:
    print(e)
```
