```python
class NotPalindromeError(Exception):
    def __init__(self):
        super().__init__('회문이 아닙니다.')  # 에러 메세지 넣기
def palindrome(word):
    if word!=''.join(reversed(word)):
        #또는 word!=word[::-1]
        raise NotPalindromeError
    else:
        print(word)
try:
    word = input()
    palindrome(word)
except NotPalindromeError as e:
    print(e)
```
