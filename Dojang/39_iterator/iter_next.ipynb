### 파이썬 내장 함수 - iter, next
- iter: 객체의 __iter__ 메서드를 호출

```python
it = iter(range(3))
print(next(it))
print(next(it))
print(next(it))
```

```python
import random

it=iter(lambda : random.randint(0,5),2)  # 0~5까지 무작위로 숫자 생성, 2가 나오면 반복 종료
print(next(it)) # 3
print(next(it)) # 1
print(next(it)) # 2여서 반복 종료, StopIteration
```

```python
for i in iter(lambda : random.randint(0,5),2):      # 이렇게 하면 if 조건문으로 2인지 확인할
    print(i, end=' ')                               # 필요가 없음
```

- next: 객체의 __next__ 메서드를 호출


```python
it=iter(range(3))
print(next(it,10))
print(next(it,10))
print(next(it,10))      # 2까지 정상적으로 나오다가
print(next(it,10))      # 기본값으로 지정했던 10이 나옴
```
