### 이터레이터(iterator)
- 값을 차례대로 꺼낼 수 있는 객체
- 연속된 숫자를 미리 만들면 메모리를 너무 많이 사용하게 됨 -> 값이 필요한 시점이 되었을 때 값을 만듦
- 지연 평가: 데이터 생성을 뒤로 미루는 것

```python
dir([1,2,3])  # dir 을 사용하면 객체의 메소드 확인 가능
              # __iter__ 메소드 존재함
```

```python
it=[1,2,3].__iter__()
it.__next__()
it.__next__()       # next로 요소를 꺼내다가, 요소가 없으면 예외 발생
```

- 반복 가능한 객체는 __iter__ 메소드로 이터레이터를 얻고, 이터레이터의 __next__ 메소드로 반복
- 반복 가능한 객체(iterable) / 이터레이터(iterator)


```python
# range 처럼 동작하는 이터레이터
class Counter:
    def __init__(self,stop):  # 초기값
        self.current=0
        self.stop=stop
    
    def __iter__(self):
        return self
    
    def __next__(self):       # 조건식, 현재값
        if self.current<self.stop:
            r=self.current
            self.current+=1
            return r
        else:
            raise StopIteration

for i in Counter(3):
    print(i,end=' ')
```

```python
# 인덱스로 접근할 수 있는 이터레이터
class Counter2:
    def __init__(self,stop):
        self.stop=stop
    
    def __getitem__(self,index):
        if index<self.stop:
            return index*2
        else:
            raise IndexError

print(Counter2(3)[0],Counter2(3)[1],Counter2(3)[2])
for i in Counter2(3):
    print(i, end=' ')
```
