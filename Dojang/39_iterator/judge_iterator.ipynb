```python
class TimeIterator:
    def __init__(self,start,stop):
        self.start=start
        self.stop=stop
    def __iter__(self):
        return ('{0}:{1}:{2}'.format)
    def __getitem__(self,index):
        if index<self.stop:
            return index
        else:
            raise IndexError

start, stop, index = map(int, input().split())
 
for i in TimeIterator(start, stop):
    print(i)
 
print('\n', TimeIterator(start, stop)[index], sep='')
```
