```python
def file_read():
    with open('words.txt') as file:
        while True:
            line=file.readline()        # 파일 한 줄씩 읽어서
            if line=='':                # 다음 줄이 없으면 멈추고
                break   
            yield line.strip('\n')      # 줄바꿈을 없앤 후 전달
for i in file_read():
    print(i)
```
