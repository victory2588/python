### 제네레이터(generator)
- 이터레이터를 생성해주는 함수
- 함수 안에 yield 라는 키워드만 사용하면 끝

```python
def number_generator():
    yield 0
    yield 1
    yield 2

for i in number_generator():    # yield에 지정했던 0, 1, 2가 나옴
    print(i)
```

```python
g=number_generator()
dir(g)  # __iter__과 __next__ 존재
```

### yield 의 동작 과정 알아보기

```python
def number_generator():
    yield 0    # 0을 함수 바깥으로 전달하면서 코드 실행을 함수 바깥에 양보
    yield 1    # 1을 함수 바깥으로 전달하면서 코드 실행을 함수 바깥에 양보
    yield 2    # 2를 함수 바깥으로 전달하면서 코드 실행을 함수 바깥에 양보
 
g = number_generator()
 
a = next(g)    # yield를 사용하여 함수 바깥으로 전달한 값은 next의 반환값으로 나옴
print(a)       # 0
 
b = next(g)
print(b)       # 1
 
c = next(g)
print(c)       # 2
```

### 제너레이터와 return

```python
def one_generator():
    yield 1
    return 'return에 지정한 값'  # 제너레이터 안에서 return 값 지정하면 
                                # StopIteration 예외의 에러 메시지로 들어감
try:
    g = one_generator()
    next(g)
    next(g)
except StopIteration as e:
    print(e)    # return에 지정한 값
```
