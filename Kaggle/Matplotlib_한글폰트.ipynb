```python
# 그래프를 노트북 안에 그리기
%matplotlib inline

# 필요한 패키지, 라이브러리
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

# 그래프에서 마이너스 폰트 그리기
mpl.rcParams['axes.unicode_minus']=False
```

```python
import numpy as np

data=np.random.randint(-100,100,50).cumsum()
data
```

```python
[(f.name,f.fname) for f in fm.fontManager.ttflist if 'Nanum' in f.name]
```

```python
plt.plot(range(50),data,'r')
mpl.rcParams['axes.unicode_minus']=False
plt.title('시간별 가격 추이')
plt.ylabel('주식 가격')
plt.xlabel('시간(분)')
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import numpy as np

data=np.random.randint(-100,100,50).cumsum()  # 임의의 값 생성

plt.plot(range(50),data,'r')
mpl.rcParams['axes.unicode_minus']=False
plt.title('시간별 가격 추이')
plt.ylabel('주식 가격')
plt.xlabel('시간(분)')
```
