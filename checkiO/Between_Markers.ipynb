```python
# 시작 마커, 끝 마커가 있는지 확인(find method 사용) -> 그 사이에 있는 문자열 가져오기
def between_marker(text:str,begin:str,end:str) -> str:
    if begin in text:
        begin_index=text.find(begin) + len(begin)
    else:
        begin_index=0
    
    if end in text:
        end_index=text.find(end)
    else:
        end_index=len(text)

    return text[begin_index:end_index]
```

```python
between_marker("What is >apple<?","<",">")
```
