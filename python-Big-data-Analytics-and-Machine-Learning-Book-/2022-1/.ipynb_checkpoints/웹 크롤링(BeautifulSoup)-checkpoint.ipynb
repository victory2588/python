>요소 1개 찾기(find)

```python
# 태그 : div 속성 : class 속성값 : temperature_info
from bs4 import BeautifulSoup as bs 
from pprint import pprint
import requests

html = requests.get('https://search.naver.com/search.naver?query=날씨')
soup = bs(html.text,'html.parser')
dustdata_one = soup.find('div',{'class':'temperature_text'}) 
print(dustdata_one)
```

>요소 모두 찾기(findAll)

```python
from bs4 import BeautifulSoup as bs 
from pprint import pprint
import requests

html = requests.get('https://search.naver.com/search.naver?query=날씨')
soup = bs(html.text,'html.parser')
dustdata_one = soup.find('div',{'class':'temperature_info'})
dustdata_all = dustdata_one.findAll('dd')  # dd 태그 찾기
pprint(dustdata_all)
```

>내부 텍스트 추출 - span - <태그>텍스트에서 텍스트만 골라내는 방법

```python
from bs4 import BeautifulSoup as bs 
from pprint import pprint
import requests

html = requests.get('https://search.naver.com/search.naver?query=날씨')
soup = bs(html.text,'html.parser')
dustdata_one=soup.find('div',{'class':'temperature_info'})
dustdata_all=dustdata_one.findAll('span')
fine_dust_code = dustdata_all[0].find('span',{'class':'blind'})
fine_dust_con = dustdata_all[0].find('span',{'class':'blind'}).text
print(fine_dust_code)
print(fine_dust_con)
```
