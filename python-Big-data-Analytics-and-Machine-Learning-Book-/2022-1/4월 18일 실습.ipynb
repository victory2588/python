# 데이터 그룹분석

```python
import pandas as pd
file_path='C:/Users/victo/anaconda3/python/실습데이터/exam_sample.csv'
df=pd.read_csv(file_path)
df
```

```python
# 반 별로 그룹화(A,B,C반)
df1=df.groupby(['class'])
df1 #이렇게만 하면 안 나오는 이유..?
df1.get_group('A')

# 반 별 그룹 평균 확인
df.groupby('class').mean()

# 반 별, 성별 그룹 평균 확인
df.groupby(['class','sex']).mean() #대괄호 꼭

# 반 별 수학 평균
df['math'].groupby(df['class']).mean()
df.groupby(df['class'])['math'].mean() #둘 다 가능

# 반별 수학 개수
df['math'].groupby(df['class']).count()

# 성별 수학 평균
df_mean=df['math'].groupby(df['sex']).mean()
sexgroup=df.groupby('sex')
print(df_mean)
print(sexgroup)
print(sexgroup.groups) #이건 뭐지?

# 남학생 수학 평균
male=sexgroup.get_group('m') #성별로 나눈 것에서 m만 뽑음
subset=male[['sex','math']] #m의 sex와 math를 뽑아서 subset 만듦
print(male)
print(subset)
```

# 데이터 재구조화

```python
# 데이터 구간화
import pandas as pd
import numpy as np 

file_path='C:/Users/victo/anaconda3/python/실습데이터/exam_sample.csv'
df = pd.read_csv(file_path)
print(df)
df1=pd.cut(df.math,3) #math 변수에 대해서 3개 동일한 길이로 범주형 변수로 만들어보자
df1

df2=df.math.groupby(df1)
df2
df3=df2.agg(['count','mean','std','min','max'])#각 범주의 그룹별로 열을 추가해서 그룹화
df3

df4=pd.qcut(df.math,3,labels=False) #math에 대해서 각 범주 구간 별로 동일하게 3개의 개수를 가지도록 범주를 만들수도 있다.
df4

df5=pd.qcut(df.math,3,labels=np.arange(3,0,-1)) #label이 3, 2, 1 로 3부터 1씩 줄어드는 순서로 할당
df5

df6=df.math.groupby(df5)
df7=df6.agg(['count','mean','std','min','max'])
df7
```

```python
# 원-핫 인코딩
df.dtypes

df=pd.get_dummies(df) # 범주형 원핫코딩
df

#데이터 전치
df.T

#피벗 테이블
pd.pivot_table(df,index='class',columns='sex',values='science') # 반(class) 변수를 행 데이터로 피봇
pd.pivot_table(df,index='sex',columns='class',values='science')

#멜트 함수
pd.melt(df,id_vars=['student_no','class'])

#스택, 언스택
df5=df.stack()
df5

df6=df.unstack(level=-1)
df6
```

```python
# 시계열 활용

import pandas as pd
file_path='C:/Users/victo/anaconda3/python/실습데이터/timeseries.csv'
df= pd.read_csv(file_path)
df['new_Date'] = pd.to_datetime(df['Date'])
df

df['Year'] = df['new_Date'].dt.year
df['Month'] = df['new_Date'].dt.month
df['Day'] = df['new_Date'].dt.day
df.head()

# to_period()함수를 이용해 표기 변경 
df['Date_yr'] = df['new_Date'].dt.to_period(freq = 'A')  # 연도까지
df['Date_m'] = df['new_Date'].dt.to_period(freq = 'M')   # 연월까지
df.head()

# 날짜 인덱스 지정
df.set_index('new_Date',inplace=True)
df.head()

# 날짜 인덱싱
print(df.loc['2015-07']) # 7월에 해당하는 row 인덱싱
print('\n')
print(df['2019-06']) # 6월에 해당하는 row 인덱싱
print('\n')
print(df['2018-06-25' : '2018-06-20'])  # 해당기간의 인덱싱

# 오늘날짜와 차리 열 추가
today = pd.to_datetime('2020-03-18')
df['time_diff'] = today - df.index
df.head()

# 데이터를 다시 불러와 시간 인덱스 지정 후 단순 시각화
df= pd.read_csv(file_path)
df['new_Date'] = pd.to_datetime(df['Date'])
df.set_index('new_Date',inplace=True)
df.plot()
```
