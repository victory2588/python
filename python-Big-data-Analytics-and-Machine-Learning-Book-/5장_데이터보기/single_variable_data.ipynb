### 변수 유형별 데이터 분석 - 범주형, 연속형
- 범주형: 빈도분석, 교차분석

```python
# 범주형
import pandas as pd
import seaborn as sns

tips=sns.load_dataset("tips")

tips
```

```python
# tips 데이터 유형 확인
tips.dtypes
```

```python
# 범주형 빈도 분석 - value_counts()
tips.value_counts('sex')
tips.value_counts('smoker')
tips.value_counts('day')
tips.value_counts('time')
```

```python
# 범주형 교차 분석 - pandas.cross_tab()
pd.crosstab(tips.sex,tips.day)  # 대괄호로 해도 됨
pd.crosstab(tips.sex,tips.day, margins=True)  # 누적값 cumulatives
pd.crosstab(tips.sex,tips.day).apply(lambda r:r/len(tips),axis=1)  # 전체 빈도 비율 확인, lambda는 어떻게 쓰는 거지?
```
