- 데이터 구간화

```python
import pandas as pd
import numpy as np

file_path='C:/python_data/7장데이터/exam_sample.csv'
df=pd.read_csv(file_path)
df
```

```python
# cut() -> 동일한 길이로 나눠서 범주형 변수 만들기
df1=pd.cut(df.math,3)  # 세 부분으로 나눈 후 그 범위 자체를 저장
df1

df2=df.math.groupby(df1)  # 위에서 정한 범위로 범위를 나눔
df3=df2.agg(['count','mean','std','min','max'])  # agg? 어쨌든 각 범위 별 산술통계
df3
```

```python
# qcut -> 동일한 개수가 들어가도록 범위 나누기   동일 개수라는게 뭐지..? 지금 개수가 다른디
df4=pd.qcut(df.math,3,labels=False)
df4

df5=pd.qcut(df.math,3,labels=np.arange(3,0,-1)) # label 이 3,2,1로 3부터 1씩 줄어드는 순서로 할당이 됨
df5

df6=df.math.groupby(df5)
df7=df6.agg(['count','mean','std','min','max'])
df7
```
