### 데이터 그룹 분석 - 집계 함수 이용

```python
import pandas as pd

file_path='C:/python_data/7장데이터/exam_sample.csv'
df=pd.read_csv(file_path)
df
```

```python
# 반 별로 그룹화
df_class=df.groupby(['class'])
df_class  # 이렇게만 하면 오브젝트만 생성됨
df_class.get_group('A')  # 이렇게 해야 나눈 그룹들 볼 수 있음
```

```python
df.groupby('class').mean()  # 그룹별 평균
df.groupby(['class','sex']).mean()

# 반 별 수학 평균
df['math'].groupby(df['class']).mean()  # math 중에서 class로 묶든지, class로 묶은 후 math를 선택하든지 똑같다?
df.groupby(df['class'])['math'].mean()

# 반 별 수학 개수
df['math'].groupby(df['class']).count()

# 성별 수학 평균
df['math'].groupby(df.sex).mean()

# 남학생 수학 평균
male=df.groupby('sex').get_group('m')
subset=male[['sex','math']]
print(male)
print(subset)
```
