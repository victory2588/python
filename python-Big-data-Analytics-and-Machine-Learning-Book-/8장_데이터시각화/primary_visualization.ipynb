### 기본 시각화

```python
import pandas as pd
import seaborn as sns

tips=sns.load_dataset("tips")
tips
```

```python
# 기본 선
tips.plot()
tips.plot(kind='line')  # plot() 속성으로 kind 지정(그래프 모양 변경 가능)
```

```python
# 히스토그램
tips.plot(kind="hist")
```

```python
# 박스플롯
tips.plot(kind="box")
```

### 주요 그래프
- 선 그래프
- 막대 그래프
- 산점도
- 히스토그램

```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# 선 그래프
tips=sns.load_dataset("tips")
plt.plot(tips.total_bill)  # 자동으로 x,y 찾아서 그래프 그림
```

```python
# 막대 그래프
plt.bar(tips.sex,tips.total_bill)
```

```python
# 산점도
plt.scatter(tips.total_bill,tips.tip)
```

```python
# 히스토그램
plt.hist(tips.total_bill)
```
