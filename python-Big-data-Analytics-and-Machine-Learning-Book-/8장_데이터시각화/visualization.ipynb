### 시각화

```python
import matplotlib as mpl
import matplotlib.pyplot as plt

plt.plot([1,2,3,4])
plt.show()
```

### 시각화 옵션 추가
- 제목: title()
- 범례: plot() 함수에 label 속성 추가
- 색상: plot() 함수에 color 속성 추가
- X,Y축 이름: plt.xlabel() / plt.ylabel()
- 그래프 선 모양: plot() 함수에 linestyle 속성 추가
- 그림 범위 지정: xlim / ylim

```python
# 제목 
plt.title('line')
plt.plot([1,2,3],[1,2,5])
plt.show()
```

```python
# 범례
plt.title('Legend')
plt.plot([1,2,3],label='asc')
plt.plot([5,4,2],label='desc')
plt.legend()
plt.show()
```

```python
# 색상
plt.plot([1,2,3],[1,3,4],color='red')
plt.show()
```

```python
# X,Y축 이름
plt.plot([1,2,3],[1,2,4])
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
# 그래프 선 모양
plt.plot([1,2,3],color='r',linestyle='--',label='dashed')
plt.plot([2,4,5],color='b',linestyle=':',label='dotted')
plt.show()
```

```python
# 그림 범위 지정
plt.title('x, y Range')
plt.plot([10,20,30,40],[1,4,9,16],c="b",lw=5,ls="--",marker="o",ms=15,mec="g",mew=5,mfc="r")  # color -> c / linestyle -> ls 등 줄이기 가능!
plt.xlim(0,50)
plt.ylim(-10,30)
plt.show()
```
