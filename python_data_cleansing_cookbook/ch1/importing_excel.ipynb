```python
import pandas as pd

percapitaGDP=pd.read_excel('C:/data-cleansing-main/Chapter01/data/GDPpercapita.xlsx',
                           sheet_name='OECD.Stat export',                               # 불러올 시트를 지정
                           skiprows=4,                                                  # 건너뛸 행 지정
                           skipfooter=1,                                                # 마지막 행 건너뜀(?)
                           usecols="A,C:T")                                             # A, C~T 열을 가지고 오도록 설정
percapitaGDP
```

### info() 로 각 열의 자료형, NaN이 아닌 행의 수 확인

```python
percapitaGDP.info()     # 각 열의 자료형, NaN이 아닌 행의 수 확인
```

### 열 이름 변경 후, 해당 열 데이터들의 공백 제거
- str.startswith / str.endswith / str.strip

```python
percapitaGDP.rename(columns={'Year':'metro'},inplace=True)
percapitaGDP['metro'].str.startswith(' ').any()             # metro 값 중 도시 이름 앞, 뒤에 공백이 추가될 수 있다. 앞을 검사하려면 startswith()
percapitaGDP['metro'].str.endswith(' ').any()               # 뒤를 검사하려면 endswith()를 사용한다. any()는 그런 행이 있는지 검사한다는 것
percapitaGDP['metro']=percapitaGDP['metro'].str.strip()     # strip으로 앞 뒤 공백 제거, 이 모든 함수는 str 안에 있는 것인 듯
```

### for 문으로 열 이름 변경 및 각 열의 데이터를 정수형으로 변환
- errors='coerce' 으로 강제 변환 가능

```python
# 데이터 열들을 숫자로 변환
for col in percapitaGDP.columns[1:]:
    percapitaGDP[col]=pd.to_numeric(percapitaGDP[col],errors='coerce')      # 강제로(coerce) float 자료형으로 변환, 변환 불가능한 것은 결측값으로 바뀜
    percapitaGDP.rename(columns={col:'pcGDP'+col},inplace=True)             # 컬럼 이름 바꾸기, 컬럼 이름이 숫자면 구문 오류로 사용하기 힘들다!
```

### 결측값 처리
- 지정된 모든 열의 값이 누락되었을 때 드롭

```python
percapitaGDP.dropna(subset=percapitaGDP.columns[1:],how='all',inplace=True)     # 지정된 열의 모든 값이 누락되었을 때 드롭한다.
percapitaGDP.shape
```
