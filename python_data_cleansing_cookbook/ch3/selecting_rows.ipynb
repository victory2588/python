## 행 선택
- 집계한 숫자 vs 특정 행 -> 둘의 중간쯤에서 드러나는 이슈도 있다
- 행들의 부분집합

```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns',None)

nls97=pd.read_csv('C:/data-cleansing-main/Chapter03/data/nls97.csv')
```

```python
nls97.set_index('personid',inplace=True)
```

```python
nls97[1000:1004].T
nls97[1000:1004:2].T        # 2는 스텝 크기, 생략할 경우 1
```

```python
# loc 사용
nls97.loc[[195884,195891,195970]].T     # 여기는 []가 2개인데
nls97.loc[195884:195970].T              # 여기는 []가 1개이다...
```

```python
# iloc 사용
nls97.iloc[[0]].T

nls97.iloc[[0,1,2]].T
nls97.iloc[0:3].T           # 슬라이싱 할 때는 []를 1개만 써도 되는걸까?

nls97.iloc[[-3,-2,-1]].T
nls97.iloc[-3:].T
```

```python
# 불리언 인덱싱으로 조건에 의한 행 선택
nls97['nightlyhrssleep'].quantile(0.05)         # quantile: 분위수에 해당하는 값 반환, 0.05의 경우 5%에 해당하는 값 반환
```

```python
nls97['nightlyhrssleep'].count()

sleepcheckbool=nls97['nightlyhrssleep']<=4

lowsleep=nls97.loc[sleepcheckbool]              # loc는 불린 값으로 특정 값들을 가져올 수 있다. 특정 조건을 만족시키는 데이터를 가져올 때 사용
lowsleep
```

```python
lowsleep['childathome'].describe()
```

```python
lowsleep3pluschildrend=nls97.loc[(nls97['nightlyhrssleep']<=4) & (nls97['nightlyhrssleep']>=3)]     # 다중 조건 검사 해보기!
lowsleep3pluschildrend
```

```python
lowsleep3pluschildrend=nls97.loc[(nls97['nightlyhrssleep']<=4) & (nls97['nightlyhrssleep']>=3),['nightlyhrssleep','childathome']]       # 다중 조건 + 특정 컬럼
lowsleep3pluschildrend
```
