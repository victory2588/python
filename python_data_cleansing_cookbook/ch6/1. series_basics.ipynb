## 판다스 시리즈에서 값 얻기
- 판다스 시리즈 구조는 1차원 배열과 유사, 넘파이 자료형을 취함
- 각 시리즈에는 인덱스가 존재!
- 시리즈 생성 방법: 리스트, 딕셔너리, 넘파이 배열, 스칼라

```python
import pandas as pd

nls97=pd.read_csv('C:/data-cleansing-main/Chapter06/data/nls97b.csv')
nls97.set_index('personid',inplace=True)
```

```python
nls97
```

```python
# 특정 열로부터 시리즈 생성
gpaoverall=nls97.gpaoverall
type(gpaoverall)
gpaoverall.head()
gpaoverall.index
```

### 대괄호 연산자

```python
gpaoverall[:5]
gpaoverall.head(5)

gpaoverall[-5:]
gpaoverall.tail(5)
```

### loc 접근자
- loc 접근자에 인덱스 레이블(여기서는 personid 값)을 전달하면 스칼라 값이 반환됨.
- 💥인덱스 레이블의 리스트를 전달하면 값이 몇 개이든 시리즈를 얻음.

```python
type(gpaoverall.loc[100061])        # 리스트x 인 경우에는 float로 
```

```python
type(gpaoverall.loc[[100061]])      # 리스트로 전달한 경우 시리즈로

gpaoverall.loc[[100061,100139,100284]]
gpaoverall.loc[100061:100833]
```

### iloc 접근자
- 행 번호의 리스트를 취함
- loc와 마찬가지로 리스트를 전달하면 시리즈로 얻음

```python
type(gpaoverall.iloc[0])
```

```python
type(gpaoverall.iloc[[0,1,2,3,4]])

gpaoverall.iloc[:5]
gpaoverall.iloc[-5:]
```
