## groupby 심화 - 체이닝 등

```python
import pandas as pd

nls97=pd.read_csv('C:/data-cleansing-main/Chapter07/data/nls97b.csv')
nls97.set_index('personid',inplace=True)
```

```python
nls97.iloc[:,0:7].info()
```

```python
# 범주형 데이터 따로!
catvars=['gender','maritalstatus','highestdegree']

for col in catvars:
    print(col,nls97[col].value_counts().sort_index(),sep="\n\n",end="\n\n\n")
```

```python
# 수치형 데이터 따로
contvars = ['satmath','satverbal','weeksworked06','gpaoverall','childathome']
nls97[contvars].describe()
```

```python
# 성별에 따른 수학점수
nls97.groupby(['gender'])['satmath'].mean()
```

```python
# 성별과 학력에 따른 수학 점수
nls97.groupby(['gender','highestdegree'])['satmath'].mean()
```

```python
# 성별과 학력에 따른 수학, 언어 점수
nls97.groupby(['gender','highestdegree'])['satmath','satverbal'].mean()
```

```python
# agg 함수를 사용해 요약통계 반환
nls97.groupby(['gender','highestdegree'])['gpaoverall'].agg(['count','mean','max','std'])
```

```python
# 딕셔너리를 사용해 복잡한 집계 수행
pd.options.display.float_format = '{:,.1f}'.format

aggdict = {'weeksworked06':['count', 'mean', 'max','std'], 'childathome':['count', 'mean', 'max', 'std']}       # 각 컬럼에 대해 다른 집계 함수를 적용할 수도 있겠네
nls97.groupby(['highestdegree']).agg(aggdict)
nls97.groupby(['maritalstatus']).agg(aggdict)
```

## 정리
- groupby를 사용할 때의 세 단계
1. 하나 이상의 범주형 변수를 기초로 groupby 데이터프레임 생성
2. 요약통계에 사용할 컬럼 선택
3. 집계 함수 선택
