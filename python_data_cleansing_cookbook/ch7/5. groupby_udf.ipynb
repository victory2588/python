## 사용자 정의 함수 및 apply

```python
import pandas as pd
import numpy as np

nls97=pd.read_csv('C:/data-cleansing-main/Chapter07/data/nls97b.csv')
nls97.set_index('personid',inplace=True)
```

```python
# 사분위범위를 정의하는 함수 작성
def iqr(x):
    return x.quantile(0.75)-x.quantile(0.25)
```

```python
aggdict={'weeksworked06':['count','mean',iqr],'childathome':['count','mean',iqr]}
nls97.groupby(['highestdegree']).agg(aggdict)
```

```python
# 선택한 요약통계를 시리즈로 반환하는 함수 정의
def gettots(x):
    out={}
    out['qr1']=x.quantile(0.25)
    out['med']=x.median()
    out['qr3']=x.quantile(0.75)
    out['count']=x.count()
    return pd.Series(out)
```

```python
pd.options.display.float_format='{:,.0f}'.format
nls97.groupby(['highestdegree'])['weeksworked06'].apply(gettots)
```

```python
# 디폴트 인덱스를 사용하기 위해 reset_index를 사용
nls97.groupby(['highestdegree'])['weeksworked06'].apply(gettots).reset_index()
```

### unstack(): 열의 축에 대한 인덱스의 일부를 회전하려고 할 때 유용(???)

```python
# 요약 변수에 기초한 열을 생성하기 위해 unstack과 엮기(??)
# highestdegree를 인덱스로 하고 집계 값들이 열이 되는 데이터프레임이 만들어짐

nlssums=nls97.groupby(['highestdegree'])['weeksworked06'].apply(gettots).unstack()      # unstack(): 인덱스 값을 컬럼으로 올려줌
nlssums
```
