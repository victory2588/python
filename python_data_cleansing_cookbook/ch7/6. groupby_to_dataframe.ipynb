## groupby를 사용해 데이터프레임의 분석 단위 바꾸기
- 분석 단위를 바꾸기 위해 데이터를 집계해야 할 때가 있음
- groupby는 분석 단위를 조정하거나 요약 작업이 필요할 때 유용
- 중복되지 않은 행들만 선택해야할 때 sort_values와 drop_duplictes 조합

```python
import pandas as pd

coviddaily=pd.read_csv('C:/data-cleansing-main/Chapter07/data/coviddaily720.csv',parse_dates=['casedate'])
ltbrazil=pd.read_csv('C:/data-cleansing-main/Chapter07/data/ltbrazil.csv')
```

```python
# 코로나 19 데이터를 국가별 일일 데이터에서 전체 국가의 일일 데이터로 변환
coviddailytotals=coviddaily.loc[coviddaily.casedate.between('2020-02-01','2020-07-12')].groupby(['casedate'],as_index=False)['new_cases','new_deaths'].sum()
```

```python
coviddailytotals
```

```python
# 브라질 기상 관측소별 평균 온도 데이터프레임 생성
ltbrazil=ltbrazil.dropna(subset=['temperature'])
ltbrazil[['station','year','month','temperature','elevation','latabs']]
```

```python
ltbrazil.groupby(['station'],as_index=False).agg({'latabs':'first','elevation':'first','temperature':'mean'})       # first는 첫 번째 값을 말하는건가?
```
