## 열 그룹 녹이기 
- 두 그룹의 열을 녹이려면? -> melt를 두 번 실행해야 했음
- wide_to_long 함수 사용하면 쉽게 녹일 수 있다. melt와 비슷하지만 기능이 많고 복잡

```python
import pandas as pd

nls97=pd.read_csv('C:/data-cleansing-main/Chapter09/data/nls97f.csv')
nls97.set_index('personid',inplace=True)
```

```python
weeksworkedcols = ['weeksworked00','weeksworked01','weeksworked02','weeksworked03','weeksworked04']
colenrcols = ['colenroct00','colenroct01','colenroct02','colenroct03','colenroct04']

nls97.loc[nls97.originalid.isin([1,2]), ['originalid']+weeksworkedcols+colenrcols]
```

### wide_to_long 사용법
> pd.wide_to_long(df, ["pred"], i="id", j="model")
1. df: pd.DataFrame
2. stubnames: 아래로 펼칠 칼럼명의 공통 문자. 이 문자로 시작하는 칼럼들을 long하게 늘여놓겠다! 이런 의미
3. i: 인덱스로 사용할 칼럼
4. j: 공통문자 다음에 오는 값이 저장될 칼럼명

```python
workschool=pd.wide_to_long(nls97[['originalid'] + weeksworkedcols + colenrcols], stubnames=['weeksworked','colenroct'],
                           i=['originalid'], j='year').reset_index()
workschool
```

```python
workschool['year']=workschool.year+2000
workschool=workschool.sort_values(['originalid','year'])
workschool.set_index(['originalid'],inplace=True)
workschool.head(10)
```
