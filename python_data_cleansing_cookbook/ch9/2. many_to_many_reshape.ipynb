```python
# 고유한 id, citation, creator 값들의 개수 확인
print(cma.id.nunique())
print(cma.citation.nunique())
print(cma.creator.nunique())
```

```python
# 중복된 id, citation, creator 값 제거
cma.drop_duplicates(['id','citation']).id.count()
cma.drop_duplicates(['id','creator']).id.count()
```

```python
# 중복된 인용 및 작성자가 있는 소장품 보이기
cma.set_index(['id'],inplace=True)
cma.loc[124733,['title','citation','creator','birth_year']].head(14)
```

```python
# 소장품 데이터프레임 생성
collectionsvars=['title','collection','type']
cmacollections=cma[collectionsvars].reset_index().drop_duplicates(['id']).set_index(['id'])
cmacollections.head()
```

```python
# 인용 데이터프레임 생성
cmacitations=cma[['citation']].reset_index().drop_duplicates(['id','citation']).set_index(['id'])       # id가 인덱스이므로 중복을 드롭하기 전에 인덱스를 먼저 리셋
cmacitations.loc[124733]
```

```python
# 작가 데이터프레임 생성
creatorsvars=['creator','birth_year','death_year']

cmacreators=cma[creatorsvars].reset_index().drop_duplicates(['id','creator']).set_index(['id'])
cmacreators.loc[124733]
```

```python
# 1950년 이후 출생한 작가의 작품 수 세기
cmacreators['birth_year']=cmacreators.birth_year.str.findall('\d+').str[0].astype(float)     # 왜 to_numeric이나 astype(float)로는 변환이 안 되는걸까..?
youngartists=cmacreators.loc[cmacreators.birth_year>1950,['creator']].assign(creatorbornafter1950='Y')
youngartists.shape[0]==youngartists.index.nunique()
youngartists
```

> 한 개의 데이터셋으로부터 소장품, 인용, 작가 데이터셋을 만들어냄
